### visualizing the data

**distributions of the variables.**

```python
data[continuous_features].hist(bins=(32),figsize=(12,15))
plt.show()
```

```python
for feature in continuous_features:
    sns.distplot(data[feature],color='red')
    plt.show()
```

```python
for feature in categorical_features:
    sns.countplot(y=data[feature])
    plt.show()
```

```python
data['Income_Category'].value_counts()
```

### Identifying Outliers

```python
for feature in continuous_features:
    sns.boxplot(data[feature])
    plt.show()
```

### Relationship between continuous features

```python
def single_scatter(data,variable_index):
    x=data.columns[variable_index]
    y=list(data.columns)
    y.pop(variable_index)
    #plt.figure(figsize=(12,8))
    for feature in y:
        data.plot(kind="scatter",x=x,y=feature)
```

```python
single_scatter(data[continuous_features],2)
```

```python
plt.figure(figsize=(16,14))
sns.heatmap(data[continuous_features].corr(),cmap='bwr',annot=True)
```

### the relationship between continuous (numerical) features and categorical (discrete) features 

```python
def map_continuous_categorical(data,continue_feature,categorical_feature):
    grid=sns.FacetGrid(data,col=categorical_feature)
    grid.map(plt.hist,continue_feature,bins=20)
```

```python
print(f"continuous features\n {data[continuous_features].columns}\n")

print(f"categorical features\n {data[categorical_features].columns}\n")
```

```python
map_continuous_categorical(data,'Customer_Age','Gender')
```

```python
def catplot(data,continue_feature,categorical_feature):
    sns.catplot(y=categorical_feature, x=continue_feature, kind="box", data=data, palette="bwr")
    plt.show()
```

```python
catplot(data,'Customer_Age','Income_Category')
```

### the relationship between categorical features

```python
def countplot(data,categorical_feature1,categorical_feature2):
    sns.countplot(data=data, y=categorical_feature1, hue=categorical_feature2, palette="bwr")
    plt.show()
```

```python
countplot(data, 'Education_Level','Gender')
```

```python
countplot(data, 'Marital_Status','Gender')
```
